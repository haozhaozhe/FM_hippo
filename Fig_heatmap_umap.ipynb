```python
col_list = ['Hoch1', 'Hoch2','JA','Harris','Zhong']
index_list = list(ad.obs['label_13'].cat.categories)
df500_name_1 = pd.DataFrame(columns = col_list , index = index_list)
df500_name_2 = pd.DataFrame(columns = col_list , index = index_list)
df500_perc_1 = pd.DataFrame(columns = col_list , index = index_list)
df500_perc_2 = pd.DataFrame(columns = col_list , index = index_list)


key1 = plot_matrix_Hoch1_500
key2 = plot_matrix_Hoch2_500
key3 = plot_matrix_JA_500
key4 = plot_matrix_harris_500
key5 = plot_matrix_zhong_500


key1_max = key1.idxmax(axis = 1)
key2_max = key2.idxmax(axis = 1)
key3_max = key3.idxmax(axis = 1)
key4_max = key4.idxmax(axis = 1)
key5_max = key5.idxmax(axis = 1)

key1_max_value = key1.max(axis = 1)
key2_max_value = key2.max(axis = 1)
key3_max_value = key3.max(axis = 1)
key4_max_value = key4.max(axis = 1)
key5_max_value = key5.max(axis = 1)


df500_name_1.loc[:,'Hoch1'] = key1_max
df500_name_1.loc[:,'Hoch2'] = key2_max
df500_name_1.loc[:,'JA'] = key3_max
df500_name_1.loc[:,'Harris'] = key4_max
df500_name_1.loc[:,'Zhong'] = key5_max

        
df500_perc_1.loc[:,'Hoch1'] = key1_max_value
df500_perc_1.loc[:,'Hoch2'] = key2_max_value
df500_perc_1.loc[:,'JA'] = key3_max_value
df500_perc_1.loc[:,'Harris'] = key4_max_value
df500_perc_1.loc[:,'Zhong'] = key5_max_value
        

for clust in list(ad.obs['label_13'].cat.categories):

    df500_name_2.loc[clust,'Hoch1'] = str(key1.loc[clust].nlargest(2).index[1])
    df500_name_2.loc[clust,'Hoch2'] = str(key2.loc[clust].nlargest(2).index[1])
    df500_name_2.loc[clust,'JA'] = str(key3.loc[clust].nlargest(2).index[1])
    df500_name_2.loc[clust,'Harris'] = str(key4.loc[clust].nlargest(2).index[1])
    df500_name_2.loc[clust,'Zhong'] = str(key5.loc[clust].nlargest(2).index[1])
    
    
    df500_perc_2.loc[clust,'Hoch1'] = str(key1.loc[clust].nlargest(2)[1])
    df500_perc_2.loc[clust,'Hoch2'] = str(key2.loc[clust].nlargest(2)[1])
    df500_perc_2.loc[clust,'JA'] = str(key3.loc[clust].nlargest(2)[1])    
    df500_perc_2.loc[clust,'Harris'] = str(key4.loc[clust].nlargest(2)[1])
    df500_perc_2.loc[clust,'Zhong'] = str(key5.loc[clust].nlargest(2)[1])
    
```
